```python
import pandas as pd
IOWA_file_path = ("/Users/shashikant/desktop/github/kaggle_learn/Machine Learning/train.csv")
data_IOWA = pd.read_csv(IOWA_file_path)
data_IOWA.describe()
```

```python
data_IOWA.head()
```

```python
#Target field for the prediction

y = data_IOWA.SalePrice
```

```python
#Create the dataframe X holding the predictive features to predict the sale price of an property
#LotArea, YearBuilt, 1stFlrSF, 2ndFlrSF, FullBath, BedroomAbvGr, TotRmsAbvGrd

feature_names = ["LotArea","YearBuilt","1stFlrSF","2ndFlrSF","FullBath","BedroomAbvGr","TotRmsAbvGrd"]
X = data_IOWA[feature_names]


```

```python
X.head()
```

```python
X.describe()
```

```python
#Specify and Fit Model

#import Decision tree regressor from sklearn.tree libraray

from sklearn.tree import DecisionTreeRegressor

#DecisionTreeRegressor with random_status = 1

IOWA_model = DecisionTreeRegressor (random_state=1)

#Fit the model

IOWA_model.fit(X,y)
```

```python
#Predictions of the first 5 rows

X.head()
predictions = IOWA_model.predict(X.head())

print(predictions)
```

```python
#Prediction of the all the data points

predictios_X = IOWA_model.predict(X)
print(predictios_X)
```

```python
# model Validation "Mean Absolute Error" (In-sample validation problem)

from sklearn.metrics import mean_absolute_error
predicted_home_price = IOWA_model.predict(X)
mean_absolute_error (y, predicted_home_price)
```

```python
# model validation using train_test_split

from sklearn.model_selection import train_test_split

#split train and val data
train_X, val_X, train_y, val_y = train_test_split(X, y, random_state=0)

#Fit the model
IOWA_model.fit(train_X, train_y)

#predicted price on validation data

val_predictions = IOWA_model.predict(val_X)
print(mean_absolute_error(val_y, val_predictions))
```

```python
#overfitting and underfitting

from sklearn.metrics import mean_absolute_error
from sklearn.tree import DecisionTreeRegressor

def get_mae(max_leaf_nodes, train_X, val_X, train_y, val_y):
    model = DecisionTreeRegressor(max_leaf_nodes=max_leaf_nodes, random_state=0)
    model.fit(train_X, train_y)
    preds_val = model.predict(val_X)
    mae = mean_absolute_error(val_y, preds_val)
    return(mae)
```
